### Qual o ritmo de crescimento da população urbana da América Latina de 1960 a 2020?

```python
df_brazil = df[df['País'].str.contains('Brazil')]
df_argentina = df[df['País'].str.contains('Argentina')]
df_uruguay = df[df['País'].str.contains('Uruguay')]
df_paraguay = df[df['País'].str.contains('Paraguay')]
df_chile = df[df['País'].str.contains('Chile')]
df_colombia = df[df['País'].str.contains('Colombia')]
df_bolivia = df[df['País'].str.contains('Bolivia')]
df_venezuela = df[df['País'].str.contains('Venezuela')]
df_guyana = df[df['País'].str.contains('Guyana')]
df_peru = df[df['País'].str.contains('Peru')]
df_ecuador = df[df['País'].str.contains('Ecuador')]
df_suriname = df[df['País'].str.contains('Suriname')] 

frames = [df_brazil, df_argentina, df_peru, df_bolivia, df_chile, df_guyana, df_ecuador, df_colombia, df_uruguay, df_paraguay, df_venezuela, df_suriname]

df_america_sul = pd.concat(frames)
df_america_sul
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(data = df_america_sul, x = 'Ano', y = 'População urbana (% da população total)', color='red').set_title('Ritmo de crescimento da população urbana na América do Sul (1960-2020)')
```

```python
filtro = [2021]
df_AS_2021 = df_america_sul[df_america_sul['Ano'].isin(filtro)]
df_AS_2021
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data = df_AS_2021, x = 'País', y = 'População urbana (% da população total)', palette = 'winter').set_title('Percentual da população urbana na América do Sul (2021)')
```
